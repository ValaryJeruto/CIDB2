#### ###Plotting a bar chart for 'Total Dewormed' for each county

```python
df['Total Dewormed'] = pd.to_numeric(df['Total Dewormed'], errors='coerce')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='county', y='Total Dewormed', data=df)
plt.xticks(rotation=90)
plt.xlabel('County')
plt.ylabel('Total Dewormed')
plt.title('Total Dewormed by County')
plt.show()
```

##### Scatter plot for 'Acute Malnutrition' vs. 'stunted 0-<6 months' for the last 100 entries

```python
df_last_100 = df.tail(100)
plt.figure(figsize=(15, 10))
sns.scatterplot(x='Acute Malnutrition', y='stunted 0-<6 months', data=df_last_100, hue='county')
plt.xlabel('Acute Malnutrition')
plt.ylabel('Stunted (0-<6 months)')
plt.title('Acute Malnutrition vs. Stunted (0-<6 months) for the Last 100 Entries')
plt.show()
```

```python
df['diarrhoea cases'] = pd.to_numeric(df['diarrhoea cases'], errors='coerce')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['period'], bins=10, alpha=0.7, color='skyblue', edgecolor='black', label='Diarrhoea Cases')
plt.xlabel('period')
plt.ylabel('Frequency')
plt.title('Histogram of Diarrhoea Cases by Period')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['period'], df['Total Dewormed'], marker='o', color='g', label='Total Dewormed')
plt.xlabel('Period')
plt.ylabel('Total Dewormed')
plt.title('Trend of Total Dewormed')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['period'], df['Acute Malnutrition'], marker='o', color='b', label='Acute Malnutrition')
plt.xlabel('period')
plt.ylabel('Acute Malnutrition')
plt.title('Trend of Acute Malnutrition')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
